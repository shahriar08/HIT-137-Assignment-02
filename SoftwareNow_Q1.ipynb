```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import string
from collections import Counter
import re

nltk.download('punkt')
nltk.download('stopwords')

# Read the content of the file
with open('all_textfile.txt', 'r') as file:
    text = file.read()

# Tokenization and initial cleaning
tokens = word_tokenize(text)
tokens = [word.lower() for word in tokens if word.strip()]

# Punctuation removal
table = str.maketrans('', '', string.punctuation)
tokens = [word.translate(table) for word in tokens if word.translate(table)]

# Stop word removal
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]

# Stemming
stemmer = PorterStemmer()
tokens = [stemmer.stem(word) for word in tokens]

# Remove digits and apply stemming
tokens = [stemmer.stem(re.sub(r'\d+', '', word)) for word in tokens]

# Remove empty strings
tokens = [word for word in tokens if word]

# Count word frequencies
word_counts = Counter(tokens)

# Get the 30 most common words
most_common_words = word_counts.most_common(30)

print(most_common_words)
```

#Model: biobert

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
import string
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Ensure required NLTK resources are downloaded
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Load BioBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-v1.1")
model = AutoModelForTokenClassification.from_pretrained("dmis-lab/biobert-v1.1")

# Read the content of the file
with open('all_textfile.txt', 'r') as file:
    text = file.read()

# Tokenization using AutoTokenizer
tokens = tokenizer.tokenize(text)

# Punctuation removal (tokenizer already handles it, but we'll do it for any remaining cases)
table = str.maketrans('', '', string.punctuation)
tokens = [word.translate(table) for word in tokens if word.translate(table) != '']

# Tokenization and initial cleaning
tokens = word_tokenize(text)
tokens = [word.lower() for word in tokens if word.strip()]

# Punctuation removal
tokens = [word.translate(table) for word in tokens if word.translate(table)]

# Stop word removal
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]

# Stemming
stemmer = PorterStemmer()
tokens = [stemmer.stem(word) for word in tokens]

# Remove digits and apply stemming
tokens = [stemmer.stem(re.sub(r'\d+', '', word)) for word in tokens]

# Remove empty strings
tokens = [word for word in tokens if word]

# Join the cleaned tokens back into a single string
cleaned_text = ' '.join(tokens)

# Create NER pipeline
nlp = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

# Process the cleaned text with BioBERT
ner_results = nlp(cleaned_text)

# Print the detected entities, combining subwords correctly
for entity in ner_results:
    print(f"Entity: {entity['word']}, Detected: {entity['entity_group']}, Confidence: {entity['score']:.3f}")
```



### 1. **`dmis-lab/biobert-v1.1`**

**Description**:
- **BioBERT (Bidirectional Encoder Representations from Transformers for Biomedical Text)** is a variant of BERT (Bidirectional Encoder Representations from Transformers) specifically pre-trained on large-scale biomedical corpora, such as PubMed abstracts and clinical notes.
- **`dmis-lab/biobert-v1.1`** is a version of BioBERT trained for general biomedical NER tasks. It captures contextual information from biomedical text, making it suitable for a wide range of biomedical named entities, such as diseases, drugs, and other medical concepts.

**Strengths**:
- **Pre-training Data**: Trained on large biomedical text datasets, which helps in understanding and generating biomedical-related language.
- **General Biomedical NER**: Suitable for a broad spectrum of biomedical entity recognition tasks.

**Limitations**:
- **Not Fine-Tuned**: It is not specifically fine-tuned on specialized datasets like chemicals or diseases, so performance on specific entity types may not be as high as specialized models.

### 2. **`en_ner_bc5cdr_md`**

**Description**:
- **`en_ner_bc5cdr_md`** is a model fine-tuned on the **BC5CDR** dataset, which is specifically annotated for biomedical concepts related to chemicals and diseases. BC5CDR is a corpus of biomedical literature with annotations for drug names, disease names, and other related entities.
- **`en_ner_bc5cdr_md`** is based on spaCy's medium-sized English model and fine-tuned to recognize entities within the BC5CDR dataset.

**Strengths**:
- **Fine-Tuned for Specific Tasks**: Specifically fine-tuned on BC5CDR, so it excels at recognizing chemicals and diseases.
- **High Accuracy**: Likely to provide better performance in identifying these specific types of entities due to its specialized training.

**Limitations**:
- **Limited Scope**: While it is highly effective for chemicals and diseases, it may not perform as well on other types of biomedical entities that are not covered by the BC5CDR dataset.

### **Which One Provides Better Results?**

**`en_ner_bc5cdr_md`** will generally provide better results for detecting **diseases** and **chemicals** specifically, due to its fine-tuning on the BC5CDR dataset which focuses on these types of entities.

- **For Chemicals and Diseases**: `en_ner_bc5cdr_md` is better suited because it is fine-tuned on a dataset specifically for these entity types.
- **For General Biomedical NER**: `dmis-lab/biobert-v1.1` is a strong general-purpose model, but it may not be as specialized in recognizing chemicals and diseases as `en_ner_bc5cdr_md`.

**In Summary**:
- Use **`en_ner_bc5cdr_md`** if your primary goal is to accurately detect chemicals and diseases.
- Use **`dmis-lab/biobert-v1.1`** if you need a more general biomedical NER model and are dealing with a broader range of biomedical entities.
